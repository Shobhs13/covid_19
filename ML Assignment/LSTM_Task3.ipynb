```python
training_size_r=int(len(df1)*0.65)
test_size_r=len(df1)-training_size_r
train_data_recovery,test_data_recovery=df1[0:training_size_r,:],df1[training_size_r:len(df1),:1]
```

```python
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]  
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
time_step = 100
X_train_rec, y_train_rec = create_dataset(train_data_recovery, time_step)
X_test_rec, y_test_rec = create_dataset(test_data_recovery, time_step)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train_rec =X_train_rec.reshape(X_train_rec.shape[0],X_train_rec.shape[1] , 1)
X_test_rec = X_test_rec.reshape(X_test_rec.shape[0],X_test_rec.shape[1] , 1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model_rec=Sequential()
model_rec.add(LSTM(50,return_sequences=False,input_shape=X_train_rec.shape))
model_rec.add(LSTM(50,return_sequences=True))
model_rec.add(LSTM(50))
model_rec.add(Dense(1))
model_rec.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model_rec.summary()
```

```python
model_rec.fit(X_train_rec,y_train_rec,validation_data=(X_test_rec,y_test_rec),epochs=100,batch_size=64,verbose=1)
```

```python
import tensorflow as tf
```

```python
train_predict_rec=model_rec.predict(X_train_rec)
test_predict_rec=model_rec.predict(y_train_rec)
```
